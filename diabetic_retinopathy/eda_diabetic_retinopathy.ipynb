```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import cv2
```

```python
# directories
data_dir = '../datasets/aptos2019/'

train_df = pd.read_csv(data_dir + 'train.csv')
test_df = pd.read_csv(data_dir + 'test.csv')

print(train_df.head())
print(test_df.head())

# shape
print(f"Train shape: {train_df.shape}")
print(f"Test shape: {test_df.shape}")
```

```python
train_df['diagnosis'].value_counts()
```

```python
# Get the value counts of the 'diagnosis' column
value_counts = train_df['diagnosis'].value_counts()

# Plot the value counts
plt.figure(figsize=(8, 6))
value_counts.plot(kind='bar')
plt.title('Train Class Distribution')
plt.xlabel('Diagnosis')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
# Example from every label
fig, ax = plt.subplots(1, 5, figsize=(15, 6))
for i in range(5):
    sample = train_df[train_df['diagnosis'] == i].sample(1)
    image_name = sample['id_code'].item()
    X = cv2.imread(f"{TRAIN_IMG_PATH}{image_name}")
    ax[i].set_title(f"Image: {image_name}\n Label = {sample['diagnosis'].item()}", 
                    weight='bold', fontsize=10)
    ax[i].axis('off')
    ax[i].imshow(X);
```
